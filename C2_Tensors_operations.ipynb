### `alpha` and `out` parameters
---

###  1. `alpha` Parameter

#### ✔ Definition

A **scaling factor** applied to one of the operands before the operation.

---

## ✔ Syntax Example

```python
torch.add(input, other, alpha=1)
```

---

#### ✔ How it works

```python
a = torch.tensor([1,2,3])
b = torch.tensor([4,5,6])

torch.add(a, b, alpha=2)
```

=> Internally:

```
a + (alpha * b)
→ a + (2 * b)
```

=> Result:

```python
[1 + 8, 2 + 10, 3 + 12]
```

---

## ✔ Used in:

* `torch.add()`
* `torch.sub()`

### Example in subtraction:

```python
torch.sub(a, b, alpha=2)
```

=> Means:

```
a - (2 * b)
```

---

#### ✔ Key Insight

`alpha` helps avoid writing extra multiplication → improves **computational efficiency**.

---

###  2. `out` Parameter

#### ✔ Definition

Specifies a **pre-allocated tensor** where the result will be stored.

---

#### ✔ Syntax Example

```python
torch.add(a, b, out=result_tensor)
```

---

#### ✔ Example

```python
result = torch.empty(3)

torch.add(a, b, out=result)
```

=> Instead of creating a new tensor, result is stored in `result`.

---

#### ✔ Why important?

* Avoids **extra memory allocation**
* Improves **performance in large computations**
* Useful in **high-performance pipelines**

---

###  3. Combined Usage

```python
result = torch.empty(3)

torch.add(a, b, alpha=2, out=result)
```

=> Operation:

```
result = a + (2 * b)
```

---

###  4. Important Notes

### ✔ Not all functions support `alpha`

* Supported: `add`, `sub`
* Not supported: `mul`, `div`

---

##### ✔ `out` must match:

* Shape
* dtype
* device

Otherwise → runtime error

---

##### ✔ Alternative: In-place operations

```python
a.add_(b)
```

=> Similar to `out`, but modifies original tensor

---

###  Conceptual Difference

| Parameter | Purpose                                  |
| --------- | ---------------------------------------- |
| `alpha`   | Scales operand before operation          |
| `out`     | Controls memory (where result is stored) |

---

###  Vocabulary Boost

* **Scaling factor** → multiplier applied before operation
* **Pre-allocated memory** → memory reserved before computation
* **In-place operation** → modifies original data
* **Memory optimization** → reducing unnecessary allocations

---

### 🔥 Practical Understanding

* Use `alpha` → when you want **weighted operations efficiently**
* Use `out` → when working with **large tensors / performance-critical code**

---

## --> How to change dtype of a tensor (already declared)
Facing a **dtype mismatch error**, which is common in PyTorch operations like `matmul`.

---

###  Problem Restated

* `a.dtype = torch.int64`
* `b.dtype = torch.float32`
* `torch.matmul(a, b)` → ❌ Error

=> Reason: **Both tensors must have the same dtype**

---

###  How to Change dtype (Core Methods)

#### ✔ 1. Using `.to()` (Most Recommended)

```python
a = a.to(torch.float32)
```

✔ Flexible (can also change device)

---

### ✔ 2. Using Type-Specific Methods

```python
a = a.float()     # → float32
a = a.int()       # → int32
a = a.long()      # → int64
a = a.double()    # → float64
```

=> These are **shortcuts (syntactic sugar)** for `.to()`

---

#### ✔ 3. Using `.type()`

```python
a = a.type(torch.float32)
```

⚠️ Older style, less preferred

---

###  Fix Your Code

```python
a = a.float()   # convert int → float

mat_mul_ab = torch.matmul(a, b)

print(mat_mul_ab)
print(a @ b)
```

---

###  Important Insight

=> PyTorch does **not automatically cast types** in operations like NumPy sometimes does.

This is called:

* **Strict type enforcement**

---

###  Best Practice

Before operations like:

* `matmul`
* `add`
* `loss functions`

✔ Always ensure:

#### ```a.dtype == b.dtype```

---

###  Pro Tip (Efficient Handling)

Instead of converting both:

```python
a = a.to(b.dtype)
```

=> This ensures compatibility dynamically

---

###  Vocabulary Boost

* **Type casting** → converting one dtype to another
* **dtype mismatch** → incompatible data types
* **Strict typing** → no implicit conversion
* **Syntactic sugar** → shortcut method

---

### 🔥 Practical Takeaway

* Use `.float()` when working with neural networks
* Keep all tensors in **float32** unless there is a strong reason not to

---

#### Question:
=> “Why deep learning prefers float32 over int” — this will connect directly to training mechanics.


````python
print(a.dtype == b.dtype)
print(f'DataType of a = {a.dtype}, b = {b.dtype}')

#changing the dype of a to int
a = a.float()
print(f'DataType of a = {a.dtype}, b = {b.dtype}')

mat_mul_ab = torch.matmul(a,b,)
print(mat_mul_ab)
print(a@b)

print(mat_mul_ab == a@b)
````

## 5. Broadcasting (Very Important Concept)

```python
a + b
```

Even if shapes differ, PyTorch tries to **auto-adjust shapes**

### Example:

```python
a = torch.tensor([[1,2,3],
                  [4,5,6]])

b = torch.tensor([1,2,3])

a + b
```

=> `b` is broadcasted across rows

---

````python
z = torch.randint(0, 5, (1,3))
print(f'z:\n{z}')
print(f'x:\n{x}')
print(x.shape)
print(f'x and z have same shape: {z.shape == x.shape}')
print(f'a+b:\n{x+z}')
print(f'\nNOTE: Tensor z is braodcased to each row of x\n')

z1 = torch.randint(0, 5, (1,4))

try:
    print(f'a+b:\n{x+z1}')
except Exception as e:
    print(f'❌ Error: {e}')
````

## 6. Aggregation Operations
**Aggregation operation:** A computation that reduces multiple tensor elements into a single value or a smaller set of values (e.g., sum, mean).


#### Sum
```torch.sum(a)```

#### Mean
```torch.mean(a.float())```

#### Max / Min
```
torch.max(a)
torch.min(a)
```


##### All Aggregation Operations:
```python
torch.sum       -> sum = total of all elements
torch.mean      -> mean = average of all elements
torch.max       -> max = maximum element value
torch.min       -> min = minimum element value
torch.argmax    -> argmax = index of maximum element
torch.argmin    -> argmin = index of minimum element
torch.prod      -> prod = product of all elements
torch.std       -> std = standard deviation (spread of data)
torch.var       -> var = variance (measure of dispersion)
torch.median    -> median = middle value after sorting
torch.mode      -> mode = most frequent element
torch.norm      -> norm = magnitude/length of tensor
torch.all       -> all = True if all elements are True
torch.any       -> any = True if at least one element is True
```


---

##### torch.opperation(tensor_name) -> attribute
##### tensor_name.opeartion() -> Method

---

````python
print(a)
a[2][2] = 1111
# Sum of all elements:
print(f'Sum of all elements: {torch.sum(a)}\t{a.sum()}')
print(f'Mean of all elements: {torch.mean(a)}\t{a.mean()}')
print(f'Maximun element: {a.max()}')
print(f'Minimum element: {a.min()}')
print(f'argmax: {torch.argmax(a)}\t{a.argmax()}')

print(a[2][2])

print(f'Median of all elements: {a.mean()}')
print(f'Satandard deviation of tensor a: {a.std()}')
print(f'Varience of tensor a: {a.var()}')
print(f'{a.all()}')
````
