## Check for GPU access with PyTorch

````python
if torch.cuda.is_available():
    device = torch.device('cuda')
    print('Using AMD Integrated GPU', torch.cuda.get_device_name(0))
else:
    device = torch.device('cpu')
    print('Using cpu as device because hardware is not compatible to use igpu')
````

````python
x = torch.ones((3,4), device=device)
print(x)
````

For PyTorch since it's capable of running compute on both GPU or CPU, it's best practice to set up device agnostic code

e.g. run on GPU if avaiable else CPU

## PyTorch GPU Acceleration on Integrated GPUs

PyTorch’s GPU acceleration is built primarily for **CUDA**, which works with **NVIDIA GPUs**.  
For integrated GPUs (iGPUs), support depends on the type:

---

### 1. Intel Integrated GPU

PyTorch does not natively support Intel iGPUs via CUDA, but you can use:

- **PyTorch with DirectML** (Windows only)  
- **PyTorch with OpenVINO** (Intel CPUs + iGPUs, cross-platform)

#### Example: PyTorch with DirectML (Windows)

```bash
pip install torch-directml
```

```python
import torch_directml

# Create a DirectML device
dml = torch_directml.device()

# Example tensor on iGPU
x = torch.ones((3, 3), device=dml)
print(x)
```

#### Example: PyTorch with OpenVINO

```bash
pip install openvino-dev[onnx,pytorch]
```

Then convert your PyTorch model to **OpenVINO IR** and run on Intel GPU.

---

### 2. AMD Integrated GPU

PyTorch can use AMD GPUs (including some iGPUs) via **ROCm** (Linux only, supported hardware required).

#### Example: PyTorch with ROCm

```bash
# Install ROCm-enabled PyTorch
pip install torch --index-url https://download.pytorch.org/whl/rocm5.6
```

```python
import torch

if torch.cuda.is_available():
    device = torch.device("cuda")  # AMD ROCm reports as 'cuda'
    print("Using AMD GPU:", torch.cuda.get_device_name(0))
else:
    device = torch.device("cpu")

x = torch.ones((3, 3), device=device)
print(x)
```

---

### 3. Apple Silicon (M1/M2) Integrated GPU

PyTorch supports Apple’s **Metal Performance Shaders (MPS)** backend.

#### Example: PyTorch with MPS

```bash
pip install torch torchvision torchaudio
```

```python
import torch

if torch.backends.mps.is_available():
    device = torch.device("mps")
    print("Using Apple MPS GPU")
else:
    device = torch.device("cpu")

x = torch.ones((3, 3), device=device)
print(x)
```

---

### 🔹 Key Points

- **NVIDIA iGPU** → Use normal CUDA (`torch.device("cuda")`)  
- **Intel iGPU** → Use **torch-directml** (Windows) or **OpenVINO**  
- **AMD iGPU** → Use **ROCm** (Linux, supported hardware)  
- **Apple iGPU** → Use **MPS backend**

#### 1. ``stack()``: Concatenates a sequence of tensors along a **new dimension**, increasing the rank of the reasult. All the inputs tensors must have the **same shape**.
Syntax: 
* torch.stack(tensors, dim=0)

* tensors → list/tuple of tensors (same shape required)
* dim → dimension index where new axis is inserted (e.g., 0, 1, -1)

````python
a = torch.tensor([1, 2, 4])
b = torch.tensor([3,4, 5])

stacked_tensor = torch.stack([a, b], dim=0)
print(stacked_tensor)

print(torch.stack([a, b], dim=1))

print(torch.stack([a, b], dim=-1))

print(torch.stack([a, b], dim=-2))
````

Syntax:

* input = tensor of prababilities (non-negative, not necessarily normalized)
* num_samples = number of indices to draw
* replacement = True (sampling with replacement) / Falsee
