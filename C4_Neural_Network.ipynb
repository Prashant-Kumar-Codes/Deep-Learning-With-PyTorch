## Data - Loading and Preparing

Data can be almost anything... in machine learning.
* Excel Spreadsheed
* Images of any kind
* Videos - Youtube, Insta, TV shows etc.
* Audio - Songs, Podcasts etc
* DNA
* Text

Machine learning is a game of two parts:

1. Get data into a numerical representation
2. Build a model to learn patterns in that numerical representation.


To showcase this, let's create some *known* data usng the linear regression formula.

We'll use a linear regression formula to make a straight line with known **parameters**.


```python
# creating known parameters
weight = 0.7
bias = 0.3

# create 
start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim=1)     # unsqueeze will add extra dimesnion 
Y = weight * X + bias

X[:5], Y[:5], len(X), len(Y)
```

### Splitting data into training/test datasets

```python
train_split = int(0.8 * len(X))
X_train, Y_train = X[:train_split], Y[:train_split]
X_test, Y_test = X[train_split:], Y[train_split:]
```

```python
print(f'Length of X_train: {len(X_train)}')
print(f'Length of Y_train: {len(Y_train)}')
print(f'Length of X_test: {len(X_test)}')
print(f'Length of Y_test: {len(Y_test)}')
```
