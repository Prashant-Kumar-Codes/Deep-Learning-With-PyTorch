````python
class Model_Seq(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        
        self.network = nn.Sequential(
            nn.Linear(num_features, 3),
            nn.Relu(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )
        
    def forward(self, featuers):
        out = self.network(features)
        return out
````
