## How a **Perceptron** is trained:
1. Input (x) → 
2. Linear Transformation (z) → 
3. Activation Function (Sigmoid) → 
4. Output (y_predicted) → 
5. Loss Calculation (Binary Cross-Entropy) → 
6. Gradient Calculation (Backpropagation via Chain Rule) →
7. Parameter Update (Gradient Descent for the next epoch)

#### Here is the step-by-step breakdown of the formulas involved in this process.

---

### Phase 1: Forward Pass (Prediction)

Before a gradient can be calculated, the network must make a prediction.

1. **Linear Transformation:** The input feature ($x$) is multiplied by the weight ($w$) and added to the bias ($b$).

$$z = w \cdot x + b$$


2. **Activation Function:** The raw value $z$ is passed through the Sigmoid activation function to squash the output into a probability between 0 and 1.

$$y_{predicted} = \sigma(z) = \frac{1}{1 + e^{-z}}$$

---

### Phase 2: Loss Calculation (Error Measurement)

Once the prediction is made, it is compared against the actual target value ($y$). For binary classification using a Sigmoid activation, the standard loss function is **Binary Cross-Entropy (BCE)**.


$$Loss (L) = - [y \cdot \log(y_{predicted}) + (1 - y) \cdot \log(1 - y_{predicted})]$$

---

### Phase 3: Gradient Calculation (Backpropagation)

To update the model, we must determine how much the weight ($w$) and the bias ($b$) contributed to the final $Loss$. This requires calculating the partial derivatives $\frac{\partial L}{\partial w}$ and $\frac{\partial L}{\partial b}$ using the **Chain Rule of Calculus** [[29:00](http://www.youtube.com/watch?v=BECZ0UB5AR0&t=1740)].

Because the perceptron is a nested mathematical function, the Chain Rule dictates that we multiply the local derivatives moving backward from the Loss to the input parameters.

**1. The Chain Rule Formula:**


$$\frac{\partial L}{\partial w} = \frac{\partial L}{\partial y_{predicted}} \times \frac{\partial y_{predicted}}{\partial z} \times \frac{\partial z}{\partial w}$$

**2. Calculating the Local Derivatives [[30:18](http://www.youtube.com/watch?v=BECZ0UB5AR0&t=1818)]:**

* **Derivative of the Loss with respect to Prediction:** 
$$\frac{\partial L}{\partial y_{predicted}} = \frac{y_{predicted} - y}{y_{predicted} \cdot (1 - y_{predicted})}$$


* **Derivative of the Sigmoid Function with respect to $z$:** 
$$\frac{\partial y_{predicted}}{\partial z} = y_{predicted} \cdot (1 - y_{predicted})$$


* **Derivative of the Linear Equation with respect to Weight ($w$):** 
$$\frac{\partial z}{\partial w} = x$$


* **Derivative of the Linear Equation with respect to Bias ($b$):** 
$$\frac{\partial z}{\partial b} = 1$$



**3. Combining the Terms (The Mathematical Cancellation) [[31:12](http://www.youtube.com/watch?v=BECZ0UB5AR0&t=1872)]:**
When you multiply the first two derivatives, the $y_{predicted} \cdot (1 - y_{predicted})$ in the numerator and denominator perfectly cancel each other out.

This drastic simplification leaves us with the final gradient formulas:

* **Gradient for Weight:** 
$$\frac{\partial L}{\partial w} = (y_{predicted} - y) \cdot x$$


* **Gradient for Bias:** 
$$\frac{\partial L}{\partial b} = (y_{predicted} - y) \cdot 1 = (y_{predicted} - y)$$

---

### Phase 4: Parameter Update (Preparing for the Next Epoch)

Once the gradients are calculated, the model uses an optimization algorithm (typically Gradient Descent) to adjust the weights in the opposite direction of the gradient. This ensures the Loss will be slightly lower in the next epoch.

A learning rate ($\eta$) controls the size of the step taken during the update.

* **Updated Weight:** 
$$w_{new} = w - \eta \cdot \frac{\partial L}{\partial w}$$


* **Updated Bias:** 
$$b_{new} = b - \eta \cdot \frac{\partial L}{\partial b}$$



Once these new parameters are established, the next epoch begins, and the entire `forward pass → loss → backward pass` cycle repeats with the updated weights.

---

### Perceptron Learning Step-By-Step with complete Gradient Calculation

| Phase | Operation | Equation |
|-------|-----------|----------|
| Forward Pass | Linear Input (z) | $z = w \cdot x + b$ |
| Forward Pass | Activation ($\hat{y}$) | $\sigma(z) = \frac{1}{1+e^{-z}}$ |
| Loss Calculation | BCE Loss | $-(y \ln \hat{y} + (1-y) \ln (1-\hat{y}))$ |
| Backward Pass | Weight Gradient | $\frac{\partial L}{\partial w} = (\hat{y} - y)x$ |
| Backward Pass | Bias Gradient | $\frac{\partial L}{\partial b} = \hat{y} - y$ |
| Update Phase | New Weight | $w - \eta \frac{\partial L}{\partial w}$ |
| Update Phase | New Bias | $b - \eta \frac{\partial L}{\partial b}$ |
